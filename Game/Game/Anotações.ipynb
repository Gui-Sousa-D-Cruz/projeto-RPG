<h1 style='font-size:1.5em;color:red;' >Lógica do dano:</h1>

```python
#Dano.
import random
pow = 0
dex = 3
con = 2
int = 4
sab = 4
car = 1
lbase = 10
life = 10 + con
hp = life
#Função que cria um d20

def d20(scss, atrib):
    d20.sucesso = scss
    d20.d20 = random.randint(1, 20)
    d20.roll = d20.d20 + atrib
    #print(scss, atrib, d20.d20, roll)
      
    #Define o nível de sucesso no d20

    if d20.roll <= 5:
        #print('Fracasso crítico!')
        d20.f_crit = True
        d20.fail = True
    elif d20.roll >= 20:
        #print('Sucesso crítico!')
        d20.s_crit = True
        d20.fail = False
    elif d20.roll >= d20.sucesso:
        #print('Sucesso')
        d20.s_crit = False
        d20.fail = False
    else:
        #print('Fracasso')
        d20.f_crit = False
        d20.fail = True


#Função que cria um d4  
  
def d4(crit):
    d4.d4 = random.randint(1, 4)
    d4.d4_2 = random.randint(1, 4)
    d4.life = life
    if crit == False:
        d4.life -= d4.d4
        d4.dano = d4.d4  
    else:
        d4.life -= d4.d4 + d4.d4_2
        d4.dano = d4.d4 + d4.d4_2

#Teste lógico que calcula o dano
    
d20() #valor minimo e atributo
if d20.fail == False:
    #Sem dano
    print('Você não sofre nenhum dano!')
elif d20.f_crit == True:
    #Dano crítico
    d4(d20.f_crit)
    print(f'Você tomou {(d4.d4 + d4.d4_2)} de dano e está agora com {d4.life} de vida')
    life -= d4.dano
else:
    #Dano 
    d4(d20.f_crit)
    print(f'Você tomou {d4.d4} de dano e está agora com {d4.life} de vida')
    life -= d4.dano
```
